```python
import pandas as pd
import datetime
import random
import calendar

def random_date(start, end):  
    """  
    生成指定范围内的随机日期  
    """  
    return start + datetime.timedelta(  
        seconds=random.randint(0, int((end - start).total_seconds())))  
  
# 定义日期范围  
start_date = datetime.datetime(2020, 1, 1)  
end_date = datetime.datetime(2023, 12, 31)  
  
# 生成10个随机日期  
random_dates = [random_date(start_date, end_date) for _ in range(10)]  

sales_USD=[400,678,345,None,345,None,987,422,None,778]
sales_PHP=[None,None,None,442,None,556,None,None,553,None]
fx_rate=[33,45,345,None,23,None,43,22,None,43]
# 打印生成的随机日期  
random_dates
date1=random_dates[2]
```

```python
#datetime.datetime(2022, 4, 1, 23, 1, 2)
first_day_in_month=datetime.datetime(date1.year,date1.month,1)
days=calendar.monthrange(date1.year,date1.month)
# last_day_in_month=datetime.datetime(date1.year,date1.month,days[1])
first_weekday=first_day_in_month.weekday()
print(first_weekday)
days_icon=[]
for i in range(first_weekday,days[1],1):
    days_icon.append(i+1)
# day_icon = range(first_weekday,days,1)
# day_icon
print(first_day_in_month)
print(date1)
print(days_icon)
print(date1.day)
if days_icon[date1.day-1]<=7:
    print("1")
elif 7<days_icon[date1.day-1]<=14:
    print("2")
elif 14<days_icon[date1.day-1]<=21:
    print("3")
elif 21<days_icon[date1.day-1]<=28:
    print("4")
elif 28<days_icon[date1.day-1]<=35:
    print("5")
elif 35<days_icon[date1.day-1]<=42:
    print("6")
```

```python
df=pd.DataFrame()
df["Sales_USD"]=sales_USD
df["Sales_PHP"]=sales_PHP
df["Fx_rate"]=fx_rate
df["Time"]=random_dates
df
# func to get the week number in a month giving date
def Week_In_Month(date1):
    first_day_in_month=datetime.datetime(date1.year,date1.month,1)
    days=calendar.monthrange(date1.year,date1.month)
    # last_day_in_month=datetime.datetime(date1.year,date1.month,days[1])
    first_weekday=first_day_in_month.weekday()
    print(first_weekday)
    days_icon=[]
    for i in range(first_weekday,days[1]+first_weekday,1):
        days_icon.append(i+1)
    t = 0
    d = days_icon[date1.day]
    if d<=7:
        t=1
    elif 7<d<=14:
        t=2
    elif 14<d<=21:
        t=3
    elif 21<d<=28:
        t=4
    elif 28<d<=35:
        t=5
    elif 35<d<=42:
        t=6
    return t
    # return int(date1.day/7)+1
```

```python
# Gives PHP value to new row if it bigger than 0 else gives USD*Fx_Rate
df["Sales_New"]=df.apply(lambda x: x["Sales_PHP"] if x["Sales_PHP"]>0 else x["Sales_USD"]*x["Fx_rate"],axis=1)
# Change long datetime type to short type only leaves Y-m-d
df["Date"]=df["Time"].apply(lambda x:x.strftime('%Y-%m-%d'))
df["WeekNum"]=df["Time"].apply(lambda x:Week_In_Month(x))
df
```
